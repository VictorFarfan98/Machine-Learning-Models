```python
def ordinary_least_squares(x, y):
    """ Compute the OLS parameter estimation of a linear model
    Args: 
        x independent variable
        y dependent variable
    Returns: (slope, intercept)
    """

    mean_x = np.mean(x)
    mean_y = np.mean(y)

    m = len(x)

    # Using the formula to calculate b1 and b2
    numer = 0
    denom = 0
    for i in range(m):
        numer += (x[i] - mean_x) * (y[i] - mean_y)
        denom += (x[i] - mean_x) ** 2
    b1 = numer / denom
    b0 = mean_y - (b1 * mean_x)
    return b1, b0

b1,b0 = ordinary_least_squares(x, y)
b1,b0
```

```python
# Plotting Values and Regression Line

max_x = np.max(x) + 100
min_x = np.min(x) - 100

# Calculating line values x and y
lin_x = np.linspace(min_x, max_x, 1000).reshape(1000,1)
lin_y = b0 + b1 * lin_x
```

```python
# Ploting Line
plt.plot(lin_x, lin_y, color='r', label='Regression Line')
# Ploting Scatter Points
plt.scatter(x, y, c='b', label='Scatter Plot',s=3)

plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
def rmse(y, y_pred):
    """ Compute Root Mean Squares Error
    """
    m = len(y)
    error = 0
    for i in range(m):
        error += (y[i] - y_pred[i]) ** 2
        
    return np.sqrt(error/m)
```

```python
y_pred = b0 + b1 * x
rmse(y, y_pred)
```

```python
def r2(y, y_pred):
    """ Compute R squared score
    """
    m = len(y)
    y_mean = np.mean(y)
    # total sum of squares
    ss_t = 0
    # total sum of squares of residuals
    ss_r = 0
    for i in range(m):
        ss_t += (y[i] - y_mean) ** 2
        ss_r += (y[i] - y_pred[i]) ** 2
    r2 = 1 - (ss_r/ss_t)
    return r2

r2(y, y_pred)
```

# Gradient descent
Based on sample code:
 
https://en.wikipedia.org/wiki/Gradient_descent#Python

```python
next_x = 6  # We start the search at x=6
gamma = 0.01  # Step size multiplier
precision = 0.000001  # Desired precision of result
max_iters = 10000   # Maximum number of iterations

# Derivative function
def df(x):
    """ Compute the derivative of f(x) = x**4 -3x**3 + 2
        Solving dF we know 9/4 (2.25) is a global minimum
    """
    return 4 * x ** 3 - 9 * x ** 2

x_values = []
step_size = []

for _i in range(max_iters):
    
    # evaluate gradient term
    current_x = next_x
    
    # update value
    # -------------
    # move gamma times in the opposite direction
    # of the derivative
    next_x = current_x - df(current_x)*gamma

    # compute step size
    step = next_x - current_x
    
    # append values for analysis
    x_values.append(current_x)
    step_size.append(step)
    
    # check if step is smaller than my required precision
    if abs(step) <= precision:
        print('Max. Precision reached')
        break
print('Iterations: {0}'.format(_i))
print("Minimum at ", next_x)
```

```python
# plot records
# ----------------------------------------
fig = plt.figure(figsize=(30,5))
ax1 = fig.add_subplot(111)
# ax1.plot(x_values[1:], c='g',label='Values')
ax1.plot(x_values, c='g',label='Values')
ax1.grid()
ax1.set_xlabel('Iterations')
ax1.set_ylabel('Value')
ax1.axhline(2.25,c='r', lw=1, ls='--')
ax1.legend(loc=3)
ax2 = ax1.twinx()
ax2.set_ylabel('Step size')
# ax2.plot(step_size[1:], c='b', label='Step size')
ax2.plot(step_size, c='b', label='Step size')
ax2.legend(loc=2)
plt.show()
```

```python
fig = plt.figure(figsize=(30,5))
ax1 = fig.add_subplot(111)
ax1.plot(x_values[1:],step_size[1:],color='b',lw=1)
ax1.scatter(x_values[1:],step_size[1:],color='red',alpha=0.8)

ax1.set_xlabel('x values')
ax1.set_ylabel('step size')
ax1.grid()
```

```python
def cost_function(X, Y, B):
    """ Compute the mean square error
    Args:
        X (np.array): Training data of shape [n_observations, n_features+1]
        Y (np.arrap): Target variable of shape [n_observations]
        B (np.array): Coeficients of shape [n_features+1]
    Returs:
        c (float): Mean square error
    
    """
    assert x.shape[0], Y.shape[0]
    assert x.shape[1], B.shape[1]
    
    n = len(Y)
    c = np.sum((X.dot(B) - Y) ** 2)/(2 * n)
    return c

print('Initial cost: {0}'.format(cost_function(X, Y, B)))
```

```python
def gradient_descent(X, Y, B, alpha, iterations):
    """ Apply gradient descent algorithm
    """
    cost = []
    gradients = []
    N = len(Y)
    
    for iteration in tqdm(range(iterations)):
        # compute predictions
        h = X.dot(B)
        # estimate loss
        loss = h - Y
        
        # gradient term calculation
        gradient = X.T.dot(loss)/N
        
        # update B values
        B = B - gradient*alpha
        
        # store current cost
        current_cost = cost_function(X, Y, B)
        cost.append(current_cost)
        gradients.append(gradient)
        
    return B, cost, gradients
```

```python
B, cost_history, grad_history = gradient_descent(X, Y, B, alpha, 1000)
```

```python
print('Coeficients: {0}'.format(B))
print('Final cost value: {0}'.format(cost_history[-1]))
```

```python
def rmse(Y, Y_pred):
    rmse = np.sqrt(sum((Y - Y_pred) ** 2) / len(Y))
    return rmse

# Model Evaluation - R2 Score
def r2_score(Y, Y_pred):
    mean_y = np.mean(Y)
    ss_tot = sum((Y - mean_y) ** 2)
    ss_res = sum((Y - Y_pred) ** 2)
    r2 = 1 - (ss_res / ss_tot)
    
    return r2

Y_pred = X.dot(B)

print(rmse(Y, Y_pred))
print(r2_score(Y, Y_pred))
```
